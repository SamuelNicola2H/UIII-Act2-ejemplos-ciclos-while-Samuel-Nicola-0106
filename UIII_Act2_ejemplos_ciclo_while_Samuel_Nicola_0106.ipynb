### Ejemplos de ciclos while

```python
# 1. Suma de números hasta un centinela negativo
# Este programa acumula la suma de valores ingresados y se detiene únicamente cuando el usuario introduce un número negativo
33
3
suma = 0  # Inicializa el acumulador de la suma en cero.
while True:  # Crea un bucle infinito que se controlará internamente.
    numero = int(input("Ingresa un número (negativo para salir): "))  # Solicita un número al usuario.
    if numero < 0:  # Verifica si el número es el centinela de salida (negativo).
        break  # Rompe el ciclo inmediatamente si la condición es verdadera.
    suma += numero  # Suma el número ingresado al acumulador total.
print(f"La suma total es: {suma}")  # Muestra el resultado final acumulado.
```

```python
# 2. Juego de adivinar el número secreto
# El ciclo se repite comparando la entrada del usuario con un valor predefinido hasta que ambos coincidan

secreto = 7  # Define el número que el usuario debe adivinar.
intento = 0  # Inicializa la variable que almacenará la entrada del usuario.
while intento != secreto:  # Mientras el intento sea diferente al secreto, el ciclo sigue.
    intento = int(input("Adivina el número secreto: "))  # Pide al usuario que ingrese un número.
    if intento != secreto:  # Si no acertó, envía un mensaje de retroalimentación.
        print("Incorrecto, intenta de nuevo.")
print("¡Felicidades, acertaste!")  # Mensaje de éxito al salir del ciclo por acertar.
```

```python
# 3. Validación de entrada (Rango de meses)
# Se utiliza para obligar al usuario a introducir un dato válido dentro de un rango específico (en este caso, del 1 al 12)

mes = 0  # Inicializa la variable mes en un valor fuera del rango válido.
while mes < 1 or mes > 12:  # Se ejecuta mientras el valor esté fuera de 1 a 12.
    mes = int(input("Introduce un número de mes (1-12): "))  # Solicita la entrada.
    if mes < 1 or mes > 12:  # Verifica si el dato es inválido para avisar al usuario.
        print("Valor fuera de rango. Inténtalo otra vez.")
print(f"Has seleccionado el mes {mes}.")  # Confirma el dato válido ingresado.
```

```python
# 4. Contador ascendente hasta un límite definido
# El usuario decide hasta qué número debe llegar el contador partiendo desde uno

limite = int(input("¿Hasta qué número quieres contar? "))  # El usuario define el final.
i = 1  # Inicializa el contador en uno.
while i <= limite:  # Ejecuta el bloque mientras el contador sea menor o igual al límite.
    print(i)  # Imprime el valor actual del contador.
    i += 1  # Incrementa el contador en una unidad en cada vuelta.
```

```python
# 5. Cálculo del factorial de un número
# Utiliza el ciclo para realizar multiplicaciones sucesivas de forma descendente basándose en un número dado por el usuario

n = int(input("Ingresa un número para calcular su factorial: "))  # Recibe el número base.
factorial = 1  # Inicializa el resultado en 1 (elemento neutro de la multiplicación).
original = n  # Guarda el valor original para el mensaje final.
while n > 0:  # Mientras el número sea mayor que cero, continúa multiplicando.
    factorial *= n  # Multiplica el acumulado por el valor actual de n.
    n -= 1  # Reduce n en uno para la siguiente iteración.
print(f"El factorial de {original} es {factorial}")  # Muestra el cálculo terminado.
```
